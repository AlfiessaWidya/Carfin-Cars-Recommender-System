```python
valid_car_name_pattern = re.compile(r'^[A-Z][a-z]+\s[A-Za-z0-9]+(\s\d{4})?$')

# Fungsi untuk memeriksa validitas nama mobil
def is_valid_car_name(car_name):
    return bool(valid_car_name_pattern.match(car_name))

# Filter nama mobil valid
df["Is Valid"] = df["Car Name"].apply(is_valid_car_name)
df_cleaned = df[df["Is Valid"]].drop(columns=["Is Valid"])

# Hasil
print("Data setelah memfilter nama mobil yang valid:")
print(df_cleaned)

df_cleaned.to_csv("cleaned_car_names.csv", index=False)
```

```python
# Drop duplicate
cars_data = df.drop_duplicates()
cars_data = cars_data.drop(columns=['Ad ID', 'Model', 'Registration city', 'Car documents', 'Assembly', 'Condition', 'Seller Location', 'Description', 'Images URL\'s', 'Car Profile'])
```

```python
# Rename kolom "KM's driven" menjadi "KMs driven"
cars_data.rename(columns={"KM's driven": "KMs driven"}, inplace=True)
```

```python
# Convert 'Price' and "KM's driven" columns to numeric, handling non-numeric values by setting errors='coerce'
cars_data['Price'] = pd.to_numeric(cars_data['Price'], errors='coerce')
cars_data["KMs driven"] = pd.to_numeric(cars_data["KMs driven"], errors='coerce')
```

```python
# Handle missing values by filling with median for numerical columns, and mode for categorical ones
for column in cars_data.columns:
    if cars_data[column].dtype == 'object':
        cars_data[column].fillna(cars_data[column].mode()[0], inplace=True)
    else:
        cars_data[column].fillna(cars_data[column].median(), inplace=True)
```

```python
exchange_rate = 56.53

# Konversi kolom 'Price' ke Rupiah
cars_data['Price (IDR)'] = cars_data['Price'] * exchange_rate

# Tampilkan hasil untuk memastikan konversi berhasil
cars_data.head()
```

```python
# Drop coloum price after convert to idr
cars_data = cars_data.drop(columns=['Formatted Car Name','Is Valid',])
```

```python
# Konversi kolom 'Price (IDR)' ke tipe numerik, jika ada nilai yang bukan angka maka akan menjadi NaN
cars_data['Price (IDR)'] = pd.to_numeric(cars_data['Price (IDR)'], errors='coerce')

# Ubah format kolom 'Price (IDR)' ke dalam format Rupiah, dengan memeriksa agar tidak ada nilai NaN
cars_data['Price (IDR)'] = cars_data['Price (IDR)'].apply(lambda x: f"Rp {x:,.0f}" if pd.notnull(x) else "N/A")

# Tampilkan hasil
cars_data.head()
```

```python
# Save the DataFrame to a JSON file
json_file_path = "cars_dataset.json"
cars_data.to_json(json_file_path, orient='records', indent=4)

print(f"Table data has been saved to {json_file_path}")
```

```python
import os
from telegram import Update
from telegram.ext import Application, CommandHandler, MessageHandler, filters, ContextTypes
from openai import OpenAI
import json

# Initialize OpenAI client
client = OpenAI(
    api_key='API KEY OPENAI'
)

# Initialize conversation history
conversation_history = {}

async def start(update: Update, context: ContextTypes.DEFAULT_TYPE):
    user_id = update.effective_user.id
    conversation_history[user_id] = []
    
    welcome_message = """
    🚗 I'm Carfin, your AI Car Recommender Bot.\n
I can help you find the perfect car based on your preferences.
Just tell me what you're looking for and I'll give you personalized recommendations!
        """
    await update.message.reply_text(welcome_message)

async def handle_message(update: Update, context: ContextTypes.DEFAULT_TYPE):
    user_id = update.effective_user.id
    user_message = update.message.text
    
    # Initialize conversation history if not exists
    if user_id not in conversation_history:
        conversation_history[user_id] = []
    
    # Add user message to history
    conversation_history[user_id].append({"role": "user", "content": user_message})
    
    # Prepare conversation prompt
    system_prompt = """You are a bilingual (Indonesian and English) car recommendation assistant.
    Your primary function is ONLY to provide car recommendations and handle car-related queries.
    
    Follow these rules strictly:
    1. When providing car recommendations, ALWAYS structure your response with these specific criteria:
        1. Car Name: [Brand] [Model] [Variant]
        2. Year: [Year]
        3. KMs driven: [Total kilometers]
        4. Fuel: [Fuel type]
        5. Transmission: [Transmission type]
        6. Description: [Detailed condition and history]
        7. Car Features: [List of key features and amenities]
        8. Price (IDR): [Price in Indonesian Rupiah]

    2. If the user's message is about cars (asking for recommendations, car features, pricing, etc):
       - Respond in the same language as the user's message
       - Provide specific car recommendations based on their needs
       - Include key features and explain why they match requirements
       - Keep responses concise but informative
    
    3. If the user's message is NOT about cars (e.g., asking about shoes, food, or other topics):
       - If user speaks in Indonesian, respond:
         "Maaf, saya hanya dapat membantu Anda dengan rekomendasi mobil. Silakan tanyakan tentang mobil yang Anda inginkan."
       - If user speaks in English, respond:
         "Sorry, I can only help you with car recommendations. Please ask about the car you are interested in."
    
    4. If the user's message is unclear:
       - Ask clarifying questions about their car preferences

    5. When users seem confused or uncertain (using phrases like "bingung", "confused", "tidak tahu", "don't know", or asking very general questions), help them by asking these guiding questions one at a time:
        In Indonesian:
        1. "Berapa budget yang Anda siapkan untuk membeli mobil?"
        2. "Mobil akan digunakan untuk keperluan apa? (keluarga, pribadi, bisnis)"
        3. "Berapa jumlah penumpang yang biasanya akan menggunakan mobil ini?"
        4. "Apakah Anda lebih memilih mobil baru atau bekas?"
        5. "Fitur apa yang paling penting bagi Anda? (keselamatan, kenyamanan, performa, hemat BBM)"
        6. "Apakah Anda mencari transmisi manual atau automatic?"
        7. etc.

        In English:
        1. "What is your budget for buying a car?"
        2. "What will the car be used for? (family, personal, business)"
        3. "How many passengers will typically use this car?"
        4. "Do you prefer a new or used car?"
        5. "What features are most important to you? (safety, comfort, performance, fuel efficiency)"
        6. "Are you looking for manual or automatic transmission?"
        7. etc.

        Only ask ONE question at a time and wait for the user's response before proceeding to the next question.
        Use the same language as the user's message.   
    """
    
    messages = [
        {"role": "system", "content": system_prompt},
        *conversation_history[user_id][-5:]  # Keep last 5 messages for context
    ]
    
    try:
        # Get response from OpenAI
        response = client.chat.completions.create(
            model="ft:gpt-4o-2024-08-06:project-ta:car-product:AWHjeoF0", # GANTI MODEL YG SUDAH DI FINE-TUNE
            messages=messages,
            temperature=0.7,
            max_tokens=500
        )
        
        assistant_response = response.choices[0].message.content
        
        # Add assistant response to history
        conversation_history[user_id].append({"role": "assistant", "content": assistant_response})
        
        await update.message.reply_text(assistant_response)
        
    except Exception as e:
        error_message = "Maaf, terjadi kesalahan. Silakan coba lagi.\nSorry, an error occurred. Please try again."
        await update.message.reply_text(error_message)
        print(f"Error: {str(e)}")

async def stop(update: Update, context: ContextTypes.DEFAULT_TYPE):
    user_id = update.effective_user.id
    conversation_history[user_id] = []
    
    stop_message = """
        Conversation has been stopped. Please use /start to begin a new conversation.
        """
    await update.message.reply_text(stop_message)

def main():
    # Initialize bot
    application = Application.builder().token('API KEY TELEGRAM').build()
    
    # Add handlers
    application.add_handler(CommandHandler("start", start))
    application.add_handler(CommandHandler("stop", stop))
    application.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, handle_message))
    
    # Start bot
    print("Bot started...")
    application.run_polling()

if __name__ == '__main__':
    main()
```

```python
nest_asyncio.apply()
logging.basicConfig(format='%(asctime)s - %(name)s - %(levelname)s - %(message)s', level=logging.INFO)
logger = logging.getLogger(__name__)

class CarRecommenderBot:
    def __init__(self):
        self.telegram_token = '7583091244:AAGpF8zKyZ7dX5DYHwidpfau_c4ppFbIajQ'
        self.client = OpenAI(
            api_key='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'
        )

        # self.model_id = "ft:gpt-4o-2024-08-06:project-ta:car-product:AWHjeoF0"
        self.model_id = "gpt-4o-08-06"

        self.cars = self.load_cars()
        self.active_sessions = set()
        self.user_context = {}
        self.last_search_criteria = {}
        self.conversation_history = {}
        self.shown_recommendations = {}
        self.system_prompts = [
            {
                "role": "system",
                "content": """You are an expert car advisor! Your personality traits are:
                    1. Knowledgeable about automotive details
                    2. Practical and insight-driven
                    3. Attentive to user's specific needs
                    4. Professional and trustworthy
                    5. Explains technical details clearly"""
            },
            {
                "role": "system",
                "content": """Carfin have 2 main languages: English and Indonesian or Bahasa. You can answer in both languages."""
            },
            {
                "role": "system", 
                "content": """Carfin only want to answer topics about car. Carfin should decline if asked about topics outside of car"""
            },
            {
                "role": "system",
                "content": """Carfin is a car recommendation bot that can recommend a reference car with user personalization."""
            }
        ]

    def _is_price_in_range(self, price_str: str, min_price: float = None, max_price: float = None) -> bool:
        try:
            price_value = float(price_str.replace('Rp', '').replace(',', '').strip())
            price_millions = price_value / 1000000
            
            if min_price is not None and price_millions < min_price:
                return False
            if max_price is not None and price_millions > max_price:
                return False
            return True
        except (ValueError, TypeError) as e:
            logging.error(f"Error converting price: {e}")
            return False

    def load_cars(self) -> List[Dict]:
        try:
            with open('cars_dataset.json', 'r', encoding='utf-8') as f:
                cars = json.load(f)
            logging.info(f"Successfully loaded {len(cars)} cars")
            return cars
        except FileNotFoundError:
            logging.error("Car database file not found")
            return []
        except json.JSONDecodeError:
            logging.error("Error decoding JSON file")
            return []

    def format_price(self, price: str) -> str:
        try:
            price_cleaned = price.replace('Rp', '').replace(',', '').strip()
            return f"IDR {float(price_cleaned):,.0f}"
        except (ValueError, TypeError):
            return "Price not available"

    def parse_features(self, features_str: str) -> List[str]:
        if isinstance(features_str, str):
            try:
                return ast.literal_eval(features_str)
            except:
                return [f.strip() for f in features_str.split('/') if f.strip()]
        elif isinstance(features_str, list):
            return features_str
        return []

    def format_car_recommendation(self, car: Dict) -> str:
        features = self.parse_features(car.get('Car Features', ''))
        features_text = '\n• '.join(features) if features else 'Not specified'
        
        return f"""
🚗 {car['Car Name']} ({car['Year']})

📝 Car Detail:
• Brands: {car.get('Make')}
• KMs Driven: {car.get('KMs driven')} km
• Fuel: {car.get('Fuel')}
• Transmission: {car.get('Transmission')}

🌟 Car Features:
• {features_text}

💰 Price (IDR): {self.format_price(car['Price (IDR)'])}
"""

    def filter_cars_by_preferences(self, transmission: str = None, min_price: float = None, max_price: float = None, year: int = None, brand: str = None, price_tolerance: float = 0.2) -> List[Dict]:
        filtered_cars = self.cars.copy()
        
        if year:
            filtered_cars = [car for car in filtered_cars if car.get('Year') == year]
            
        if transmission:
            filtered_cars = [car for car in filtered_cars if car.get('Transmission', '').lower() == transmission.lower()]
            
        if brand:
            filtered_cars = [car for car in filtered_cars if car.get('Make', '').lower() == brand.lower()]
        
        if min_price is not None or max_price is not None:
            min_price_with_tolerance = min_price * (1 - price_tolerance) if min_price else None
            max_price_with_tolerance = max_price * (1 + price_tolerance) if max_price else None
            
            filtered_cars = [car for car in filtered_cars 
                           if self._is_price_in_range(car['Price (IDR)'], min_price_with_tolerance, max_price_with_tolerance)]

        if min_price or max_price:
            target_price = (min_price + max_price) / 2 if min_price and max_price else (min_price or max_price)
            filtered_cars.sort(key=lambda x: (
                abs(float(x['Price (IDR)'].replace('Rp', '').replace(',', '').strip()) / 1000000 - target_price),
                -len(self.parse_features(x.get('Car Features', '')))
            ))
        else:
            filtered_cars.sort(key=lambda x: len(self.parse_features(x.get('Car Features', ''))), reverse=True)

        return filtered_cars

    def parse_user_preferences(self, user_input: str) -> Dict:
        preferences = {}
        input_lower = user_input.lower()
        
        brand_patterns = {
            'honda': 'Honda', 'toyota': 'Toyota', 'daihatsu': 'Daihatsu',
            'suzuki': 'Suzuki', 'kia': 'KIA', 'hyundai': 'Hyundai',
            'chevrolet': 'Chevrolet', 'mitsubishi': 'Mitsubishi'
        }
        
        for brand_key, brand_value in brand_patterns.items():
            if brand_key in input_lower:
                preferences['brand'] = brand_value
                break
        
        year_patterns = [
            (r'tahun\s+([12][0-9]{3})', 'explicit year'),           
            (r'thn\s+([12][0-9]{3})', 'abbreviated year'),          
            (r'[\s,.]([12][0-9]{3})[\s,.]', 'standalone year'),     
            (r'([12][0-9]{3})\s*(?:manual|automatic|matic|at|mt)', 'year with transmission')
        ]
        
        for pattern, _ in year_patterns:
            year_match = re.search(pattern, input_lower)
            if year_match:
                try:
                    year = int(year_match.group(1))
                    if 1900 <= year <= 2024:
                        preferences['year'] = year
                        break
                except ValueError:
                    continue
        
        transmission_patterns = {
            'automatic': ['automatic', 'auto', 'at', 'a/t', 'matic', 'otomatis'],
            'manual': ['manual', 'mt', 'm/t', 'stick']
        }
        
        for trans_type, patterns in transmission_patterns.items():
            if any(pattern in input_lower for pattern in patterns):
                preferences['transmission'] = trans_type.capitalize()
                break
        
        price_patterns = [
            (r'(\d+)(?:\s*-\s*|\s+sampai\s+|\s+to\s+)(\d+)\s*(?:juta|jt|million)', 'range'),
            (r'(?:sekitar|about|around|approximately|kisaran)\s*(\d+)\s*(?:juta|jt|million)', 'approximate'),
            (r'(?:budget|harga|price)\s*(\d+)\s*(?:juta|jt|million)', 'single'),
            (r'(\d+)\s*(?:juta|jt|million)', 'simple')
        ]
        
        for pattern, pattern_type in price_patterns:
            price_match = re.search(pattern, input_lower)
            if price_match:
                try:
                    if pattern_type == 'range':
                        preferences['min_price'] = float(price_match.group(1))
                        preferences['max_price'] = float(price_match.group(2))
                    else:
                        target_price = float(price_match.group(1))
                        preferences['min_price'] = target_price * 0.8
                        preferences['max_price'] = target_price * 1.2
                    break
                except (ValueError, IndexError):
                    continue
        
        return preferences

    async def get_ai_response(self, user_input: str, user_id: int) -> str:
        try:
            history = self.conversation_history.get(user_id, [])
            messages = self.system_prompts + history + [
                {"role": "user", "content": user_input}
            ]
            
            response = self.client.chat.completions.create(
                model=self.model_id,
                messages=messages,
                temperature=0.7,
                max_tokens=1000
            )
            
            return response.choices[0].message.content
            
        except Exception as e:
            logging.error(f"Error getting AI response: {str(e)}")
            return None

    async def get_recommendation(self, user_input: str, user_id: int) -> str:
        try:
            history = self.conversation_history.get(user_id, [])
            shown_cars = self.shown_recommendations.get(user_id, set())
            
            asking_for_alternatives = any(phrase in user_input.lower() 
                for phrase in ['opsi lain', 'rekomendasi lain', 'alternatif', 'pilihan lain', 'yang lain', 'mobil lain'])
            
            asking_for_best = any(phrase in user_input.lower() 
                for phrase in ['terbaik', 'best', 'paling bagus', 'rekomendasi terbaik', 'yang terbaik'])
            
            current_preferences = self.parse_user_preferences(user_input)
            last_criteria = self.last_search_criteria.get(user_id, {})
            merged_preferences = {**last_criteria, **current_preferences}
            
            self.last_search_criteria[user_id] = merged_preferences
            
            filtered_cars = self.filter_cars_by_preferences(
                transmission=merged_preferences.get('transmission'),
                min_price=merged_preferences.get('min_price'),
                max_price=merged_preferences.get('max_price'),
                year=merged_preferences.get('year'),
                brand=merged_preferences.get('brand')
            )
            
            if not filtered_cars:
                return self.handle_no_results(merged_preferences)
            
            if asking_for_alternatives:
                filtered_cars = [car for car in filtered_cars 
                               if car['Car Name'] not in shown_cars]
            
            if asking_for_best:
                filtered_cars = self.sort_cars_by_best_match(filtered_cars, merged_preferences)
                num_recommendations = 2
            else:
                num_recommendations = 3
            
            cars_to_show = filtered_cars[:num_recommendations]
            
            if not cars_to_show:
                return "Sorry, I've already shown you all the recommendations that match your criteria."
            
            response = "Yeayy! Here are Carfin's recommendations:\n\n" if not asking_for_alternatives else "Here are some alternative recommendations:\n\n"
            
            for car in cars_to_show:
                response += self.format_car_recommendation(car)
                shown_cars.add(car['Car Name'])
            
            if len(filtered_cars) > len(cars_to_show):
                response += "\n\nThere are other suitable recommendations. Type 'other recommendations' to see more options."
            
            self.shown_recommendations[user_id] = shown_cars
            history.append({"role": "user", "content": user_input})
            history.append({"role": "assistant", "content": response})
            self.conversation_history[user_id] = history[-10:]
            
            return response
            
        except Exception as e:
            logging.error(f"Error getting recommendation: {str(e)}")
            return "Sorry, an error occurred. Please try again."
        
    def sort_cars_by_best_match(self, cars: List[Dict], preferences: Dict) -> List[Dict]:
        """Sort cars by best match considering multiple criteria"""
        def calculate_match_score(car):
            score = 0
            
            # Feature completeness (30% weight)
            features_count = len(self.parse_features(car.get('Car Features', '')))
            score += 0.3 * (features_count / 20)  # Assuming max 20 features
            
            # Price match (30% weight)
            if preferences.get('min_price') and preferences.get('max_price'):
                target_price = (preferences['min_price'] + preferences['max_price']) / 2
                price_value = float(car['Price (IDR)'].replace('Rp', '').replace(',', '').strip()) / 1000000
                price_diff_ratio = abs(price_value - target_price) / target_price
                score += 0.3 * (1 - min(price_diff_ratio, 1))
            
            # Mileage score (20% weight)
            kms_driven = float(car.get('KMs driven', '0').replace(',', ''))
            mileage_score = 1 - (kms_driven / 200000)  # Assuming 200,000 km as max
            score += 0.2 * max(0, mileage_score)
            
            # Year score (20% weight)
            if preferences.get('year'):
                year_diff = abs(car.get('Year', 2000) - preferences['year'])
                year_score = 1 - (year_diff / 5)  # 5 years difference max
                score += 0.2 * max(0, year_score)
            
            return score
        
        return sorted(cars, key=calculate_match_score, reverse=True)

    def handle_no_results(self, preferences: Dict) -> str:
        """Handle case when no cars match the criteria"""
        return (
            "Sorry, I didn't find a car that matches your criteria.\n"
            f"Your Preference:\n"
            f"• Brands: {preferences.get('brand', 'Not specified')}\n"
            f"• Year: {preferences.get('year', 'Not specified')}\n"
            f"• Transmission: {preferences.get('transmission', 'Not specified')}\n"
            f"• Price Range: {preferences.get('min_price', 'Not specified')} - "
            f"{preferences.get('max_price', 'Not specified')} million\n\n"
            "Maybe you want to:\n"
            "1. Expanding the price range\n"
            "2. Trying a different year\n"
            "3. Trying other brands\n"
            "Please customize your preferences."
        )

    async def start(self, update: Update, context: ContextTypes.DEFAULT_TYPE) -> None:
        """Send a message when the command /start is issued."""
        user = update.effective_user
        user_id = user.id
        
        # Activate session
        self.active_sessions.add(user_id)
        
        welcome_message = (
            f"Hi {user.first_name}! 🚗 I'm Carfin, your AI Car Recommender Bot.\n\n"
            "I can help you find the perfect car based on your preferences. "
            "Just tell me what you're looking for and I'll give you personalized recommendations!"
        )
        await update.message.reply_html(welcome_message)

    async def stop(self, update: Update, context: ContextTypes.DEFAULT_TYPE) -> None:
        """Clear user data and end session when /stop command is issued."""
        user_id = update.effective_user.id
        
        # Clear all user data
        if user_id in self.user_context:
            del self.user_context[user_id]
        if user_id in self.last_search_criteria:
            del self.last_search_criteria[user_id]
        if user_id in self.conversation_history:
            del self.conversation_history[user_id]
        if user_id in self.shown_recommendations:
            del self.shown_recommendations[user_id]
            
        # Remove from active sessions
        self.active_sessions.discard(user_id)
        
        await update.message.reply_text(
            "Session ended. All conversation data has been deleted. "
            "Please type /start to start a new conversation!"
        )

    async def handle_message(self, update: Update, context: ContextTypes.DEFAULT_TYPE) -> None:
        """Handle incoming messages and generate car recommendations."""
        user_id = update.effective_user.id
        user_input = update.message.text

        # Check if session is active
        if user_id not in self.active_sessions:
            await update.message.reply_text(
                "Sorry, your session has ended. "
                "Please type /start to start a new conversation!"
            )
            return

        # Process message only if session is active
        await update.message.reply_chat_action(action='typing')

        try:
            recommendation = await self.get_recommendation(user_input, user_id)
            await update.message.reply_text(recommendation)
        except Exception as e:
            logging.error(f"Error processing message: {e}")
            await update.message.reply_text("Maaf, terjadi kesalahan. Silakan coba lagi.")

    async def run(self):
        """Run the bot"""
        application = Application.builder().token(self.telegram_token).build()

        # Register handlers
        application.add_handler(CommandHandler("start", self.start))
        application.add_handler(CommandHandler("stop", self.stop))
        application.add_handler(MessageHandler(filters.TEXT & ~filters.COMMAND, self.handle_message))

        logging.info("Bot started. Press Ctrl+C to stop.")
        await application.run_polling(drop_pending_updates=True)

def main():
    """Main function to run the bot"""
    bot = CarRecommenderBot()
    try:
        asyncio.run(bot.run())
    except KeyboardInterrupt:
        print("Bot stopped by user.")
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == '__main__':
    main()
```
